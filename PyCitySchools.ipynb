```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import numpy as np
# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete.school_name.unique()
school_count
```

```python
school_countnum = len(school_count)
school_countnum
```

```python
# Calculate the total number of students
student_count = len(school_data_complete)
student_count
```

```python
# Calculate the total budget
budgets = school_data_complete.budget.unique()
total_budget = sum(budgets)
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete.math_score.mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete.reading_score.mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame([[school_countnum,student_count,total_budget,average_reading_score,average_math_score,passing_math_percentage,
                                 passing_reading_percentage,overall_passing_rate]],columns=['Total Schools','Total Students','Total Budget','Avg Math Score',
                                                      'Avg Reading Score','Passing Math %','Passing Reading %',
                                                       'Overall Passing %'])
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Avg Math Score"] = district_summary["Avg Math Score"].map("{:,.2f}".format)
district_summary["Avg Reading Score"] = district_summary["Avg Reading Score"].map("{:,.2f}".format)
district_summary["Passing Math %"] = district_summary["Passing Math %"].map("{:,.2f}".format)
district_summary["Passing Reading %"] = district_summary["Passing Reading %"].map("{:,.2f}".format)
district_summary["Overall Passing %"] = district_summary["Overall Passing %"].map("{:,.2f}".format)
# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete.groupby('school_name').type.unique()
school_types
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby('school_name').size()
#per_school_counts = school_data_complete["school_name"].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data_complete.groupby('school_name').budget.unique()
per_school_capita = per_school_budget / per_school_counts 

 
per_school_budget.astype(float)
per_school_capita.astype(float)
```

```python
# Calculate the average test scores per school
per_schooldf = school_data_complete[["school_name", "math_score", "reading_score"]]
per_schooldf 
per_school_math = per_schooldf.groupby('school_name')['math_score'].mean()

per_school_reading = per_schooldf.groupby('school_name')['reading_score'].mean()
#Send to DF to merge 
per_school_math = per_school_math.to_frame()
per_school_reading = per_school_reading.to_frame()
per_school_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passingm = per_schooldf[per_schooldf['math_score'] > 70]
students_passing_math = students_passingm.groupby('school_name')['math_score'].count()
students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passingr = per_schooldf[per_schooldf['reading_score'] > 70]
students_passing_reading = students_passingr.groupby('school_name')['reading_score'].count()
students_passing_reading 
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading 
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = students_passing_math / per_school_counts * 100
per_school_passing_reading = students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
overall_passing_rate
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = per_school_math.merge( per_school_reading, on="school_name")
per_school_summary.index.names = ['School Name']
per_school_summary = per_school_summary.rename(columns={'math_score': 'Avg Math Score', 'reading_score': 'Avg Reading Score'}) 
per_school_summary['Total Students'] = per_school_counts
per_school_summary['Total School Budget'] = per_school_budget.astype(float)
per_school_summary['Per Student Budget'] = per_school_capita.astype(float)
per_school_summary['Students Passing Math'] = students_passing_math
per_school_summary['Students Passing Reading'] = students_passing_reading
per_school_summary['Students Passing Math & Reading'] = school_students_passing_math_and_reading
per_school_summary['Passing Math %'] = per_school_passing_math
per_school_summary['Passing Reading %'] = per_school_passing_reading
per_school_summary['Overall Passing %'] = overall_passing_rate
per_school_summary['School Type'] = school_types


# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
per_school_summary["Total Students"] = per_school_summary["Total Students"].map("{:,}".format)
per_school_summary["Avg Math Score"] = per_school_summary["Avg Math Score"].map("{:,.2f}".format)
per_school_summary["Avg Reading Score"] = per_school_summary["Avg Reading Score"].map("{:,.2f}".format)
per_school_summary["Passing Math %"] = per_school_summary["Passing Math %"].map("{:,.2f}".format)
per_school_summary["Passing Reading %"] = per_school_summary["Passing Reading %"].map("{:,.2f}".format)
per_school_summary["Overall Passing %"] = per_school_summary["Overall Passing %"].map("{:,.2f}".format)

# Display the DataFrame
per_school_summary
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Per Student Budget"] = per_school_capita
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"].astype(float), bins=spending_bins, labels=labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
print(school_spending_df['Avg Reading Score'].dtype)
school_spending_df["Avg Reading Score"] = pd.to_numeric(school_spending_df["Avg Reading Score"], downcast="float")
school_spending_df["Avg Math Score"] = pd.to_numeric(school_spending_df["Avg Math Score"], downcast="float")
school_spending_df["Passing Math %"] = pd.to_numeric(school_spending_df["Passing Math %"], downcast="float")
school_spending_df["Passing Reading %"] = pd.to_numeric(school_spending_df["Passing Reading %"], downcast="float")
school_spending_df["Overall Passing %"] = pd.to_numeric(school_spending_df["Overall Passing %"], downcast="float")

#pd.to_numeric(school_spending_df['Avg Reading Score'])
#pd.to_numeric(school_spending_df['Avg Math Score'])
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Avg Math Score"].mean()
spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)")["Avg Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Passing Math %"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Passing Reading %"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing %"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame()
spending_summary['Avg Math'] = spending_math_scores
spending_summary['Avg Reading'] = spending_reading_scores
spending_summary['Passing Math'] = spending_passing_math
spending_summary['Passing Reading'] = spending_passing_reading
spending_summary['Passing Overall'] = overall_passing_spending
# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.
per_school_summary['Total Students'] = per_school_counts
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"].astype(int), bins=size_bins, labels=labels)
per_school_summary
```

```python
# Calculate averages for the desired columns. 
per_school_summary["Avg Reading Score"] = pd.to_numeric(per_school_summary["Avg Reading Score"], downcast="float")
per_school_summary["Avg Math Score"] = pd.to_numeric(per_school_summary["Avg Math Score"], downcast="float")
per_school_summary["Passing Math %"] = pd.to_numeric(per_school_summary["Passing Math %"], downcast="float")
per_school_summary["Passing Reading %"] = pd.to_numeric(per_school_summary["Passing Reading %"], downcast="float")
per_school_summary["Overall Passing %"] = pd.to_numeric(per_school_summary["Overall Passing %"], downcast="float")
size_math_scores = per_school_summary.groupby(["School Size"])["Avg Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Avg Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["Passing Math %"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["Passing Reading %"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["Overall Passing %"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame()
size_summary['Avg Math'] = size_math_scores
size_summary['Avg Reading'] = size_reading_scores
size_summary['Passing Math'] = size_passing_math
size_summary['Passing Reading'] = size_passing_reading
size_summary['Passing Overall'] = size_overall_passing
# Display results

# Display results
size_summary
```
